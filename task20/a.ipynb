```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with synthetic data
np.random.seed(42)
data = {
    'num_var1': np.random.rand(100),
    'num_var2': np.random.rand(100) * 2,
    'num_var3': np.random.rand(100) * 3
}
df = pd.DataFrame(data)

# Assuming 'df' is your DataFrame with columns 'num_var1', 'num_var2', and 'num_var3'
sns.set(style="ticks")  # Optional: Set the style for the plot, 'ticks' adds grid lines

# Create the PairGrid with the specified variables
g = sns.PairGrid(data=df, vars=['num_var1', 'num_var2', 'num_var3'])

# Map a histogram to the diagonal plots
g.map_diag(plt.hist)

# Map scatter plots to the off-diagonal plots
g.map_offdiag(plt.scatter)

# Show the plot matrix
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with numerical data
data = {
    'Age': [25, 30, 22, 40, 35, 28, 32, 37, 29, 31],
    'Income': [50000, 60000, 40000, 80000, 70000, 55000, 65000, 75000, 58000, 63000],
    'Savings': [2000, 3000, 1500, 6000, 5000, 2500, 4500, 5500, 3500, 4000]
}
df = pd.DataFrame(data)

# Assuming 'df' is your DataFrame with columns 'Age', 'Income', and 'Savings'
sns.set(style="ticks")  # Optional: Set the style for the plot, 'ticks' adds grid lines

# Create the PairGrid with the specified variables
g = sns.PairGrid(data=df, vars=['Age', 'Income', 'Savings'])

# Map a histogram to the diagonal plots
g.map_diag(plt.hist)

# Map scatter plots to the off-diagonal plots
g.map_offdiag(plt.scatter)

# Show the plot matrix
plt.show()
```
