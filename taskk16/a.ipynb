```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('ZF_Elec_Responses.xlsx')
data.head()
```

```python
data.info()
```

```python
data['Full Name (3 names at least)'] = data['Full Name (3 names at least)'].str.replace('محمد', 'Mohamed')
data['Full Name (3 names at least)'] = data['Full Name (3 names at least)'].str.replace('Mohammed', 'Mohamed')
data['Full Name (3 names at least)'] = data['Full Name (3 names at least)'].str.replace('احمد', 'Ahmed')
data['Full Name (3 names at least)'] = data['Full Name (3 names at least)'].str.replace('أحمد', 'Ahmed')

# Print the updated data
print(data)
```

```python
timestamp=data['Timestamp']
days=timestamp.dt.date
cdays=days.value_counts().sort_index()
# plt.style.use('_mpl-gallery')
fig,ax=plt.subplots(figsize=(6, 4))
ax.bar(cdays.index,cdays.values)
# plt.plot(cdays.index,cdays.values,marker='o' ,alpha=0.5)
plt.xticks(rotation=90);
```

```python
hour=timestamp.dt.hour
hours=hour.value_counts().sort_index()
fig,ax=plt.subplots(figsize=(6,4))
ax.bar(hours.index,hours.values)
```

```python
full_names = data['Full Name (3 names at least)']
# print(full_names.shape)
first_names = full_names.str.split().str[0]
name_counts = first_names.value_counts()
fig,ax=plt.subplots(figsize=(8,4))
ax.bar(name_counts.index,name_counts.values)

plt.xlabel('First Name')
plt.ylabel('Count')
plt.title('Occurrences of First Names')
plt.xticks(rotation=90);

```

```python

# mohamed_data = first_names[first_names.str.contains('Mohamed', case=False)]

mohamed_data = data[data['Full Name (3 names at least)'].str.split().str[0].str.contains('Mohamed', case=False)]
# print(mohamed_data)
average_occurrences = mohamed_data.shape[0] / data.shape[0]
print( mohamed_data.shape[0], data.shape[0],average_occurrences)
# # Extract the hour from the Timestamp column
hours = mohamed_data['Timestamp'].dt.hour
# # Count the occurrences for each hour
hour_counts = hours.value_counts().sort_index()
fig,ax=plt.subplots(figsize=(8,4))
plt.bar(hour_counts.index, hour_counts.values)
```

```python
data['Phone Number'] = data['Phone Number'].str.replace(r'^\+20', '0',regex=True)
data['Phone Number'] = data['Phone Number'].str.replace(r'^1', '01',regex=True)
print(data['Phone Number'][12] )
data['First Three Digits'] = data['Phone Number'].astype(str).str[:3]

# Count the occurrences of each first three digits
digit_counts = data['First Three Digits'].value_counts()
fig,ax=plt.subplots(figsize=(8,4))
ax.bar(digit_counts.index,digit_counts.values)
```

```python
data.to_csv('new.csv')
data.to_excel('new.xlsx')
```
